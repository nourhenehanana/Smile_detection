## Data Preprocessing

```python
df=pd.read_csv("emotion.csv")
```

```python
df.head()
```

```python
data2 = df.copy()

# Iterate over the rows of the DataFrame
for index, row in tqdm(df.iterrows(), total=len(df)):
    if row["deepface_emotion_happy_detection"] == "happy":
        data2.at[index, "deepface_emotion_happy_detection"] = 1
    else:
        data2.at[index,"deepface_emotion_happy_detection"]=0
        
    
```

```python
data2.to_csv("modified_emotions_after_test.csv")
```

## Calculat the accuracies

```python
#Calulate accuracy of the model
S=0
for index, row in tqdm(dataframe.iterrows(), total=len(dataframe)):
    if row["lenet_cnn_smile_detection"] == row["real_smile_detection"]:
        S+=1
        print("index of image",index)
        print("Equality =",S)
predicted_smile_detection_lenet_accuracy=S/(len(dataframe))
print("accuracy of our model",predicted_smile_detection_lenet_accuracy)
```

```python
#Calulate accuracy of deepface
S=0
for index, row in tqdm(dataframe.iterrows(), total=len(dataframe)):
    if row["deepface_emotion_happy_detection"] == row["real_smile_detection"]:
        S+=1
        print("index of image",index)
        print("Equality =",S)
deepface_accuracy=S/(len(dataframe))
print("accuracy of deepface",deepface_accuracy)
```
